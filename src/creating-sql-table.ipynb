```python
# Importing the necessary library
# Creating and connecting to a database where we will keep the results from the varying PD tournaments carried out

import sqlalchemy as sa
engine = sa.create_engine('sqlite:///C:/Users/sophi/Documents/final-project/source-code/Experiment-Database.db')
connection = engine.connect()
```

```python
# 'create_table' variable contains SQL code which creates our results table with the required entry fields

create_table = """
CREATE TABLE folk_theorem (
    experiment       INTEGER NOT NULL,
    player_name         TEXT NOT NULL,
    characteristics  TEXT NOT NULL,
    p_value          REAL NOT NULL,
    payoff_matrix    BLOB NOT NULL,
    repetitions      INTEGER NOT NULL,
    equilibria       BLOB NOT NULL,
    lowest_defection REAL NOT NULL,
    noise            REAL NOT NULL,
        
    CONSTRAINT folk_theorem_pk PRIMARY KEY (experiment, player),
    CONSTRAINT p_value_ck CHECK (p_value BETWEEN 0 AND 1),
    CONSTRAINT low_defect_ck CHECK (lowest_defection BETWEEN 0 AND 1)
)
"""
connection.execute(create_table)
```

```python
# code which could potentially write the results to file 

read_into_sql = """
INSERT into folk_theorem (experiment, player, characteristics, p_value, payoff_matrix, repetitions, equilibria, lowest_defection, noise)
VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?)
"""

from collections import namedtuple
result = namedtuple('result', 'experiment, player, characteristics, p_value, payoff_matrix, repetitions, equilibria, lowest_defection, noise')

# values = (foo.experiment, foo.player, foo.characteristics, foo.p_value, foo.payoff_matrix, foo.repetitions, foo.equilibria, foo.lowest_defection, foo.noise)
# connection.execute(read_into_sql, values)
```

```python
import numpy as np


pay_matrix = np.array([[3, 0], [5, 1]])
eq = [(1, 1), (1, 1)]

record = result('A', 'Tit for Tat', 'memory_depth 1', p_value=0.4, payoff_matrix=pay_matrix, repetitions=500, equilibria=eq, lowest_defection=0.2, noise=0)
values = (record.experiment, record.player, record.characteristics, record.p_value, record.payoff_matrix, record.repetitions, record.equilibria, record.lowest_defection, record.noise)
connection.execute(read_into_sql, values)
```
